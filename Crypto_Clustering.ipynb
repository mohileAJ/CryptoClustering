```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt 

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_cols = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data

df_market_data_scaled = pd.DataFrame(
    scaled_cols,
    columns=df_market_data.columns
)

# Copy the crypto names from the original data
df_market_data_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")


# Display sample data
display(df_market_data_scaled.head(5))
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_range = [k for k in range(1, 12)]
```

```python
# Create an empty list to store the inertia values
inertias = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_range:
    model = KMeans(n_clusters=k)
    model.fit(df_market_data_scaled[['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d','price_change_percentage_200d','price_change_percentage_1y']])
    inertias.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = [{"k": k, "inertia": inertia} for k, inertia in zip(k_range, inertias)]

# Create a DataFrame with the data to plot the Elbow curve
elbow_data_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_data_df.hvplot.line(x="k", y="inertia")
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** I would pick between 4 and 5.  Beyond that, we would be adding more variables without getting a significant increase in clustered density.  Further, it would make the model harder to interpret without additional benefit. 

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
K_optimal = 4
model = KMeans(n_clusters=K_optimal)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data

predictions = model.predict(df_market_data_scaled)
# Print the resulting array of cluster values.
print(predictions)
```

```python
# Create a copy of the DataFrame
df_market_data_scaled_copy = df_market_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scaled_copy["cluster"] = predictions

# Display sample data
display(df_market_data_scaled_copy.head(5))
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_data_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    color="cluster",
    hover_cols=["coin_id"]
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_cols = pca.fit_transform(df_market_data_scaled)

# View the first five rows of the DataFrame. 
display(pca_cols[:5])
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is 89.5%, of which PC1 and PC2 explain comparable ballpark range at 37.2% and 34.7% respectively, while PC3 at 17.6% is also explaining a significant level of the variance. 

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(
  pca_cols,
  columns=["PC1", "PC2", "PC3"]
)
# Creating a DataFrame with the PCA data
print(pca_df.head())

# Copy the crypto names from the original data
pca_df["coin_id"] = df_market_data.index

# Set the coinid column as index
pca_df = pca_df.set_index("coin_id")

# Display sample data
display(pca_df.head())
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_range = [k for k in range(1, 12)]
```

```python
# Create an empty list to store the inertia values
inertias = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_range:
  model = KMeans(n_clusters=k)
  model.fit(pca_df)
  inertias.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = [{"k": k, "inertia": inertia} for k, inertia in zip(k_range, inertias)]

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_data_df = pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_data_df.hvplot.line(x="k", y="inertia")
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Based on the PCA data, the inertia is lower throughout (initial inertia at 237 vs 287 for the original data),  indicating that the clusters are more tightly grouped around their respective means.
  Further, the dimensionality reduction performed by the PCA method may not have caused a significant loss of variance, given we have 90% of the variance being explained by three features (PC1, PC2, PC3). 
  
  Therefore, we do not see the need to increase the k using the PCA data, and it may be alright to maintain it at 4.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
K_optimal = 4
model = KMeans(n_clusters=K_optimal)
```

```python
# Fit the K-Means model using the PCA data
model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
predictions = model.predict(pca_df)

# Print the resulting array of cluster values.
print(predictions)
```

```python
# Create a copy of the DataFrame with the PCA data
pca_df_copy = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_df_copy["cluster"] = predictions

# Display sample data
display(pca_df_copy.head(5))
```
